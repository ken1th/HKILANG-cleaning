```r
# Define a function that rearranges the values in a row
rearrange_row <- function(row) {
  row[order(is.na(row))]
}

# Apply the function to each row of the data frame
data <- apply(df, 1, rearrange_row)
```

```r
data <- as.data.frame(data %>% t) %>% drop_na(2) %>% filter(!V1 %in% "序號") %>% select(V1:V8)
dim(data)
head(data)
data[4, 5]
```

```r
data %>% filter(!is.na(V8))
data <- data %>% select(V1:V7)
```

```r
colnames(data) <- c("Index", "Glossary", "EnglishGlossary", "Hakka", "TinhKok", "TungPingChau", "Waitau")
```

```r
data2 <- data %>% separate(Hakka, into = c("Hakka_1", "Hakka_2", "Hakka_3"), sep = "/") %>% 
    separate(TinhKok, into = c("TinhKok_1", "TinhKok_2"), sep = "/") %>% 
    separate(TungPingChau, into = c("TungPingChau_1", "TungPingChau_2"), sep = "/") %>%  
    separate(Waitau, into = c("Waitau_1", "Waitau_2", "Waitau_3"), sep = "/")
```

```r
data2[2182,7]
```

```r
data3 <- lapply(data2, function(x) {
    x <- gsub("[[:cntrl:]]+", " ", x)
    x <- gsub("^\\s+", "", x)
    x <- gsub("\\s+", " ", x)
    x <- gsub("---", NA, x)
}) %>% as.data.frame

head(data3)
data3[4, 8]
data3[8, 4]
```

```r
data3[2182,7]
```

```r
lapply(colnames(data3[4:13]), function(x){
    newcol <- c(x, paste0(x, "_p"))
    data3 %>% separate(all_of(x), into = newcol, sep = " ", extra = "merge") %>% select(all_of(newcol)) %>%
        .[!is.na(.[, 1]) & is.na(.[, 2]), ]
})
```

```r
data4 <- lapply(colnames(data3[4:13]), function(x){
    y <- data3 %>% separate(all_of(x), into = c("glossary", "phonetics"), sep = " ", extra = "merge") %>% mutate(phonetics = gsub("\\s+", "", phonetics)) %>% select(phonetics)    
    colnames(y) <- paste0(x, "_p")
    return(y)
}) %>% bind_cols
```

```r
head(data4)
write.csv(data4, "phonetics.csv", row.names = F)
```

```r
num_spaces <- sapply(glossaryDf$Hakka %>% na.omit, function(x) {
  # Replace spaces with an empty string
  x <- gsub("\\s+", " ", x)
  y <- gsub("\\s+", "", x)
  y <- gsub("/", "", x)
  
  # Subtract the number of characters after replacing spaces from the original number of characters
  return(nchar(x, type = "chars") - nchar(y))
})

# Print the resulting vector
unique(num_spaces)

glossaryDf$Hakka %>% na.omit %>% .[num_spaces >= 1]
```
