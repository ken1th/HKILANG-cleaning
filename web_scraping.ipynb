```python
data = []

for i in range(2185): 
    response = requests.get('http://www.hkilang.org/v2/word_dictionary_result/?id=' + str(i+1).zfill(4))
    soup = BeautifulSoup(response.content, "html.parser")
    tag_object = soup.find(id = 'main')
    texts = [response.status_code]
    for tag in tag_object.div.div.find(class_='fusion-four-fifth four_fifth fusion-layout-column fusion-column spacing-yes last').find_all(class_ = 'fusion-one-third one_third fusion-layout-column fusion-column spacing-yes'):
        texts.append(tag.get_text(strip = True))
    data.append(texts)

df = pd.DataFrame(data)
```

```python
import csv
from unidecode import unidecode

# Open the input CSV file
with open('data.csv', 'r', encoding='utf-8') as csv_in:
  # Create a CSV reader
    reader = csv.reader(csv_in)
  
  # Open the output CSV file
    with open('data2.csv', 'w', encoding='utf-8') as csv_out:
    # Create a CSV writer
        writer = csv.writer(csv_out)
    
    # Iterate over the rows in the input CSV file
        for row in reader:
      # Remove diacritic characters from each cell in the row
            row = [unidecode(cell) for cell in row]
      
      # Write the modified row to the output CSV file
            writer.writerow(row)
```
